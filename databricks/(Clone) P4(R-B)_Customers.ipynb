```python
df_raw_customers = (
    spark.read.csv(raw_path_customers, header=True)
    .withColumn("raw_file_path", F.col("_metadata.file_path"))
    .withColumn("raw_ingestion_timestamp", F.current_timestamp())
)
```

```python
df_bronze_customers = (
    df_raw_customers
    .withColumnRenamed("CustomerID", "customer_id")
    .withColumnRenamed("CustomerName", "customer_name")
    .withColumnRenamed("Email", "email")
    .withColumnRenamed("Phone", "phone")
    .withColumnRenamed("Address", "address")
)
```

```python
df_bronze_customers = (
    df_raw_customers
    .withColumnRenamed("CustomerID", "customer_id")
    .withColumnRenamed("FirstName", "first_name")
    .withColumnRenamed("LastName", "last_name")        # already snake_case
    .withColumnRenamed("email", "email")                # already snake_case
    .withColumnRenamed("phone", "phone")                # already snake_case
    .withColumnRenamed("city", "city")                  # already snake_case
    .withColumnRenamed("state", "state")                # already snake_case
    .withColumnRenamed("SignupDate", "signup_date")
)

df_bronze_customers.show()
```

```python
bronze_path_customers = "<location>"

df_bronze_customers.write.format("delta").mode("overwrite").save(bronze_path_customers)
```

```python
df_bronze_customers_check = spark.read.format("delta").load(bronze_path_customers)
df_bronze_customers_check.show()
```
