```python
%sql
WITH recent_orders AS (
    SELECT 
        order_id,
        customer_id,
        order_date,
        ROW_NUMBER() OVER (PARTITION BY customer_id ORDER BY order_date DESC) AS rn
    FROM delta.`<location>`
)
SELECT *
FROM recent_orders
WHERE rn = 1;
```

```python
%sql
SELECT 
    c.customer_id,
    c.first_name,
    c.last_name,
    SUM(f.total_amount) AS total_spent,
    AVG(SUM(f.total_amount)) OVER () AS avg_spent_all_customers
FROM delta.`<Location>` c
LEFT JOIN delta.`<Location>` f
    ON c.customer_id = f.customer_id
GROUP BY c.customer_id, c.first_name, c.last_name;
```

```python
%sql
SELECT 
    p.product_id,
    p.product_name,
    SUM(f.quantity) AS total_sold,
    RANK() OVER (ORDER BY SUM(f.quantity) DESC) AS sales_rank
FROM delta.`<Location>` p
RIGHT JOIN delta.`<Location>` f
    ON p.product_id = f.product_id
GROUP BY p.product_id, p.product_name;
```

```python
%sql
SELECT 
    c.customer_id,
    c.first_name,
    f.order_id,
    DENSE_RANK() OVER (ORDER BY f.order_date) AS order_sequence
FROM delta.`<Location>` c
FULL OUTER JOIN delta.`<Location>` f
    ON c.customer_id = f.customer_id;
```

```python
%sql
SELECT 
    c.customer_id,
    p.product_id
FROM delta.`<Location>` c
CROSS JOIN delta.`<Location>` p;
```

```python
%sql
SELECT
    order_id,
    order_date,
    total_amount,
    SUM(total_amount) OVER (
        ORDER BY order_date
        ROWS BETWEEN 2 PRECEDING AND CURRENT ROW
    ) AS rolling_3_order_sum
FROM delta.`<Location>`;
```

```python
%sql
SELECT
    customer_id,
    order_id,
    order_date,
    total_amount,
    LAG(total_amount, 1) OVER (
        PARTITION BY customer_id
        ORDER BY order_date
    ) AS previous_order_amount
FROM delta.`<location>`;
```
